```python
img = cv2.imread("happy.jpg", 1)
```

### Drawing a rectangle across the face

```python
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = faceCascade.detectMultiScale(gray,1.1,1)

for(x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 9)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
faces = faceCascade.detectMultiScale(gray,1.1,50)

for(x, y, w, h) in faces:
    cv2.rectangle(img1, (x, y), (x+w, y+h), (0, 255, 0), 2)

plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)) 
```

```python
img2 = cv2.imread("face.jpg", 1)
gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
faces = faceCascade.detectMultiScale(gray,1.1,50)

for(x, y, w, h) in faces:
    cv2.rectangle(img2, (x, y), (x+w, y+h), (0, 255, 0), 2)

plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)) 
```
